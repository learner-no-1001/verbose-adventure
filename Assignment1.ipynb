```python
# prompt: Ques 1. Write the code for Local Binary Pattern in python (without using its
# library function) based on the following steps.
# Ques 2. Generate the output image using this code for input images
# 1. Take the input image and convert it into a grayscale image. return grayscale image
# 2. Use a sliding window technique to traverse the input image.
# 3. For each pixel location as center, check its nearest 8 neighborhood
# 4. Threshold the neighbourhood values based on the value at the
# centre pixel and generate the 8 bit binary.
# 5. Convert 8 bit binary into decimal for every pixel as centre

import cv2
import numpy as np
from google.colab.patches import cv2_imshow # For displaying images in Colab

def calculate_lbp(image_path):
    # 1. Load and convert to grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Error: Could not load image at {image_path}")
        return None, None, None, None

    cv2_imshow(img) #display grayscale image
    print("Grayscale Image displayed")


    height, width = img.shape
    lbp_image = np.zeros((height, width), dtype=np.uint8)

    # 2 & 3 & 4 & 5. Sliding window and LBP calculation
    for i in range(1, height - 1):
        for j in range(1, width - 1):
            center_pixel = img[i, j]
            binary_code = ""

            #Check 8 neighbors
            neighbors = [img[i - 1, j - 1], img[i - 1, j], img[i - 1, j + 1],
                        img[i, j + 1], img[i + 1, j + 1], img[i + 1, j],
                        img[i + 1, j - 1], img[i, j - 1]]

            #Thresholding:Here the intensity value of the center pixel is used as a threshold.
            for neighbor in neighbors:
                if neighbor >= center_pixel:
                    binary_code += "1"
                else:
                    binary_code += "0"

            # Convert binary to decimal
            lbp_value = int(binary_code, 2)
            lbp_image[i, j] = lbp_value

    cv2_imshow(lbp_image)
    print("LBP Image displayed")

    return img, lbp_image


# Upload your image to the colab workspace before running the code.
grayscale_img, lbp_result = calculate_lbp('/content/Capture.PNG')
grayscale_img, lbp_result = calculate_lbp('/content/Capture1.PNG')
```
